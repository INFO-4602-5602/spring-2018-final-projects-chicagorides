```python
wickets = defaultdict(int)

for i,row in df.iterrows():
    if row["dismissal_kind"]!="":
        wickets[row["bowler"]]+=1
```

```python
for key, value in sorted(wickets.iteritems(), key=lambda (k,v): (v,k)):
    print ("%s: %s" % (key, value))
```
