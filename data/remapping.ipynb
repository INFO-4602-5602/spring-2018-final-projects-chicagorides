```python
import pandas as pd
import json
import numpy as np
```

```python
df =pd.read_csv("chicago_taxi_trips_2016_01.csv")
```

```python
with open('column_remapping.json') as json_data:
    js = json.load(json_data)
    
```

```python
# for row in df.iterrows():
df.columns
```

```python
import math
count = 0

change_params = ["pickup_latitude","pickup_longitude","dropoff_latitude","dropoff_longitude"]
for param in change_params:
    print("Current Param {}".format(param))
    for i in range(len(df)):
        try:
            if not math.isnan(df.at[i,param]):
                df.at[i,param] = js[param][str(int(df.at[i,param]))]
        except (KeyError,ValueError):
            print("[Error] Reading : {} and value of {} : {}".format(i, param , df.at[i,param]))
```
